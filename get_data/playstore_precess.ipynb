```python
from collections import Counter
from konlpy.tag import Okt

# 형태소 분석기 초기화
okt = Okt()

# 'rating'가 5인 행들의 'review' 열을 하나의 문자열로 결합
text = ' '.join(selected_df[selected_df['rating'] == 5]['review'])

# 형태소 분석 후 동사, 명사, 형용사 추출
pos_tags = okt.pos(text, norm=True, stem=True)
nouns_adj_verbs = [word for word, pos in pos_tags if pos in ['Noun', 'Adjective', 'Verb']]

# Counter를 사용하여 각 품사의 빈도 계산
counter = Counter(nouns_adj_verbs)

# 각 품사에서 가장 빈도가 높은 단어 10개 추출
top_nouns = counter.most_common(10)
print("명사 상위 10개:", top_nouns)

# 형용사 추출
adjectives = [word for word, pos in pos_tags if pos == 'Adjective']
adjective_counter = Counter(adjectives)
top_adjectives = adjective_counter.most_common(10)
print("형용사 상위 10개:", top_adjectives)

# 동사 추출
verbs = [word for word, pos in pos_tags if pos == 'Verb']
verb_counter = Counter(verbs)
top_verbs = verb_counter.most_common(10)
print("동사 상위 10개:", top_verbs)
```
